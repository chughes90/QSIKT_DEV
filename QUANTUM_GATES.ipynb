```python
#IMPORTS
import qiskit
import numpy as np
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_vector
```

```python
#THIS TUTORIAL IS DEVOTED TO A SUMMARY OF QUANTUM GATES
```

```python
#X (not) GATE ACTING ON |0>
x_gate = QuantumCircuit(1,1) #make a QC with 1 quantum bit and 1 classical bit

x_gate.x(0) #add a x qate to bit 0

x_gate.measure(0,0) #measure the 0th qbit and map to the 0th classical bit

x_gate.draw(output='mpl') #draw the circuit
```

```python
print('Results for X measurement:')
counts = execute(x_gate,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts) #do the simulation by getting the Aer backend end, and plot the histogram results, remember the default shots is 1024
```
