```python
#IMPORTS
import qiskit
import numpy as np
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_vector
```

```python
qc = QuantumCircuit(2) #make a QC with 2 quantum bits both initialized to |0>
qc.draw(output='mpl') #draw the circuit
```

```python
qc.h(0)
qc.cx(0,1) #control qbit 0 and target qbit 1
qc.draw(output='mpl') #draw the circuit
```

```python
cr = ClassicalRegister(2) #make a classical register to map the measurement of both qbits to
qc.add_register(cr)
qc.draw(output='mpl') #draw the circuit
```

```python
qc.measure(0,0) #MEASURE THE QUANTUM CIRCUIT (COLLAPSE THE WAVEFUNCTION) BY MAPPING TO CLASSICAL BITS
qc.measure(1,1) #MEASURE THE QUANTUM CIRCUIT (COLLAPSE THE WAVEFUNCTION) BY MAPPING TO CLASSICAL BITS
```

```python
qc.draw(output='mpl') #draw the circuit
```

```python
emulator = Aer.get_backend('qasm_simulator') #GET THE SIMULATOR BACKEND (NO NOISE) 

job = execute( qc, emulator, shots=100000 ) #SIMULATE THE QC DRAWN ABOVE FOR 100000 ITERATIONS

hist = job.result().get_counts() #GET THE RESULT OF THE SIMULATOR AND PRINT 
print(hist)
```

```python
plot_histogram( hist )
```
