```python
#IMPORTS
import qiskit
import numpy as np
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_vector
```

```python
#cx GATE ACTING ON two qbits
qc = QuantumCircuit(2,2) #make a QC with 2 quantum bits and 2 classical bits 

qc.cx(0,1) #Add a CX or controlled not gate. 0th qbit (right most) is control and 1st qbit is the target

qc.measure(0,0) #MEASURE THE QUANTUM CIRCUIT (COLLAPSE THE WAVEFUNCTION) BY MAPPING TO CLASSICAL BITS
qc.measure(1,1)

qc.draw(output='mpl') #draw the circuit
```

```python
emulator = Aer.get_backend('qasm_simulator') #GET THE SIMULATOR BACKEND (NO NOISE)

job = execute( qc, emulator, shots=100000 ) #SIMULATE THE QC DRAWN ABOVE FOR 8192 ITERATIONS
```

```python
hist = job.result().get_counts() #GET THE RESULT OF THE SIMULATOR AND PRINT 
plot_histogram( hist )
```

```python
#The above is a perfect copy, it mapped |0> onto |0> proucing |00>
```

```python
#cx GATE ACTING ON two qbits
qc = QuantumCircuit(2,2) #make a QC with 2 quantum bits and 2 classical bits 

qc.h(0) #h gates rotates into the entangled state (rotates 90 degrees about y)
qc.cx(0,1) #Add a CX or controlled not gate. 0th qbit (right most) is control and 1st qbit is the target

qc.measure(0,0) #MEASURE THE QUANTUM CIRCUIT (COLLAPSE THE WAVEFUNCTION) BY MAPPING TO CLASSICAL BITS
qc.measure(1,1)

qc.draw(output='mpl') #draw the circuit
```

```python
emulator = Aer.get_backend('qasm_simulator') #GET THE SIMULATOR BACKEND (NO NOISE)

job = execute( qc, emulator, shots=100000 ) #SIMULATE THE QC DRAWN ABOVE FOR 8192 ITERATIONS

hist = job.result().get_counts() #GET THE RESULT OF THE SIMULATOR AND PRINT 
plot_histogram( hist )
```

```python
#The above is a NOT perfect copy, it created the bell state (1/sqrt(2))*(|00> + |11>)
```
