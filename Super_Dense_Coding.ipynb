```python
#IMPORTS
import qiskit
import numpy as np
from qiskit import *
from qiskit.circuit import Gate
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_vector
```

```python
#one qbit encoding, if message is 1, use X, otherwise do not use X
qc = QuantumCircuit(2,2) #make a quantum circuit with 2 qbit and 2 classical bits
qc.h(0) #h gate on ALICEs QBIT
qc.cx(0,1) #cx with 0th qbit a control and 1st qbit as target
qc.x(0) #x gate on ALICEs QBIT
qc.measure(0,0) #ALICE qbit mapped to classical bit 0
qc.measure(1,1) #BOB qbit mapped to classical bit 1
qc.draw(output='mpl') #draw the circuit
```

```python
emulator = Aer.get_backend('qasm_simulator') #GET THE SIMULATOR BACKEND (NO NOISE)

job = execute( qc, emulator, shots=100000 ) #SIMULATE THE QC DRAWN ABOVE FOR 8192 ITERATIONS

hist = job.result().get_counts() #GET THE RESULT OF THE SIMULATOR AND PRINT 
plot_histogram( hist )
```
